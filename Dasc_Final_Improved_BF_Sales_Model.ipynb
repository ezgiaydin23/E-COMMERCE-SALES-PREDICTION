```python
##Upon looking at the data, we see that the features such as Gender, Age, Occupation, City_Category, Stay_in_Current_City_Years, Marital_Status and other Product_Category features are categorical features and should be one-hot encoded to proceed further.
##Features like user id product id and 70% null valued category 3 are not useful for the model, so we are going to drop them from the data set
##Based on the previous observations, moving forward we will be working with features Age,Gender,Occupation,City_Category,Stay in current City years, Marital Status, Product category 1.
##In order to make data cleaning and encoding easier to apply both datasets, we are defining functions.
```

```python
def DataCleaning(df):
    
    #replace
    df['Age'] = df['Age'].apply(lambda x : str(x).replace('55+', '55'))
    df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].apply(lambda x : str(x).replace('4+', '4'))
    
    #drop
    df.drop('User_ID', axis = 1, inplace = True)
    df.drop('Product_ID', axis = 1, inplace = True)
    df.drop('Product_Category_3', axis = 1, inplace = True)
    #df.drop('Product_Category_2', axis = 1, inplace = True)
    
    ##fill missing values
    df['Product_Category_2'].fillna(df['Product_Category_2'].median(), inplace = True)
    
    ##type change
    
    df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].astype(int) 
    

    
    return df
```

```python
def Feature_Encode(df):
     ##encoding
   
    df['Gender'] = np.where(df['Gender']=='M', 1, 0)

    # Age is currently grouped, but it is ordinal, so we are Applying guided ordinal encoding to map different range of ages to integer values in increasing order of age
    df['Age'] = df['Age'].map({'0-17':1, '18-25':2, '26-35':3, '36-45':4, '46-50':5, '51-55':6, '55':7})

    #other categorical features are encoded via get_dummies method.
    #df = pd.get_dummies(df, columns=['Occupation', 'City_Category','Product_Category_1'])
    df = pd.get_dummies(df,columns=['Occupation'],prefix=["Occupation"])
    df = pd.get_dummies(df,columns=['City_Category'],prefix=["City"])
    df = pd.get_dummies(df,columns=['Product_Category_1'],prefix=["Product_Category_1"])
    df = pd.get_dummies(df,columns=['Product_Category_2'],prefix=["Product_Category_2"])
    return df
```

```python
train_df_final = DataCleaning(train_df)
train_df_final = Feature_Encode(train_df_final)
```
